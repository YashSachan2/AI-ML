```python
# Define the final prediction function using the trained XGBoost model
def predict_shipment_status(shipment_details):
    # Convert shipment details into a DataFrame for easier manipulation
    shipment_data = pd.DataFrame([shipment_details], columns=[
        'Shipment ID', 'Origin', 'Destination', 'Shipment Date', 'Vehicle Type', 'Distance (km)', 'Weather Conditions', 'Traffic Conditions'
    ])

    # Drop Shipment ID (it's not relevant for prediction)
    shipment_data = shipment_data.drop(columns=['Shipment ID'])

    # Convert the 'Shipment Date', 'Planned Delivery Date', and 'Actual Delivery Date' to datetime
    shipment_data['Shipment Date'] = pd.to_datetime(shipment_data['Shipment Date'], errors='coerce')
    # shipment_data['Planned Delivery Date'] = pd.to_datetime(shipment_data['Planned Delivery Date'], errors='coerce')
    # shipment_data['Actual Delivery Date'] = pd.to_datetime(shipment_data['Actual Delivery Date'], errors='coerce')

    # Create new features for Original Diff and Actual Diff
    # shipment_data['original_diff'] = (shipment_data['Planned Delivery Date'] - shipment_data['Shipment Date']).dt.days
    # shipment_data['actual_diff'] = (shipment_data['Actual Delivery Date'] - shipment_data['Planned Delivery Date']).dt.days
    shipment_data['year'] = shipment_data['Shipment Date'].dt.year

    # Drop the original 'Planned Delivery Date', 'Shipment Date', and 'Actual Delivery Date' columns
    shipment_data = shipment_data.drop(columns=['Shipment Date'])

    # Convert categorical variables to dummy variables using pd.get_dummies
    shipment_data = pd.get_dummies(shipment_data, columns=['Origin', 'Destination', 'Vehicle Type', 'Weather Conditions', 'Traffic Conditions'], drop_first=True)

    # Define the expected column order based on the model's training data
    model_columns = [
        'Distance (km)', 'year',
        'Origin_Ahmedabad', 'Origin_Bangalore', 'Origin_Chennai', 'Origin_Delhi',
        'Origin_Hyderabad', 'Origin_Jaipur', 'Origin_Kolkata', 'Origin_Lucknow',
        'Origin_Mumbai', 'Origin_Pune', 'Destination_Ahmedabad', 'Destination_Bangalore',
        'Destination_Chennai', 'Destination_Delhi', 'Destination_Hyderabad', 'Destination_Jaipur',
        'Destination_Kolkata', 'Destination_Lucknow', 'Destination_Mumbai', 'Destination_Pune',
        'Vehicle Type_Container', 'Vehicle Type_Lorry', 'Vehicle Type_Trailer',
        'Vehicle Type_Truck', 'Weather Conditions_Clear', 'Weather Conditions_Fog',
        'Weather Conditions_Rain', 'Weather Conditions_Storm', 'Traffic Conditions_Heavy',
        'Traffic Conditions_Light', 'Traffic Conditions_Moderate'
    ]

    # Add missing columns with 0 if they are not present after encoding
    missing_cols = set(model_columns) - set(shipment_data.columns)
    for col in missing_cols:
        shipment_data[col] = 0  # Add missing columns with default value of 0

    # Reorder the columns to match the expected order
    shipment_data = shipment_data[model_columns]

    # Convert the final DataFrame to integers (XGBoost works better with integer values)
    shipment_data = shipment_data.astype(int)

    # Convert the final DataFrame to a numpy array
    final_features = shipment_data.values

    # Make prediction using the trained XGBoost model
    prediction = xgb_model.predict(final_features)

    # Return the prediction result (Delayed or On Time)
    if prediction == 1:
        return "Delayed"
    else:
        return "On Time"

```

```python
from flask import Flask, request, jsonify
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
import joblib
from pyngrok import ngrok
from threading import Thread

# Initialize Flask app
app = Flask(__name__)

# Load the trained AdaBoost model (ensure you have saved your model using joblib)
# Load the model from the file where it was saved
model_path = 'adaboost_model.pkl'
xgb_model = joblib.load(model_path)

# Define the final prediction function using the trained AdaBoost model
def predict_shipment_status(shipment_details):
    # Convert shipment details into a DataFrame for easier manipulation
    shipment_data = pd.DataFrame([shipment_details], columns=[
        'Shipment ID', 'Origin', 'Destination', 'Shipment Date', 'Vehicle Type', 'Distance (km)', 'Weather Conditions', 'Traffic Conditions'
    ])

    # Drop Shipment ID (it's not relevant for prediction)
    shipment_data = shipment_data.drop(columns=['Shipment ID'])

    # Convert the 'Shipment Date' to datetime
    shipment_data['Shipment Date'] = pd.to_datetime(shipment_data['Shipment Date'], errors='coerce')

    # Create new features for Year
    shipment_data['year'] = shipment_data['Shipment Date'].dt.year

    # Drop the original 'Shipment Date' column
    shipment_data = shipment_data.drop(columns=['Shipment Date'])

    # Convert categorical variables to dummy variables using pd.get_dummies
    shipment_data = pd.get_dummies(shipment_data, columns=['Origin', 'Destination', 'Vehicle Type', 'Weather Conditions', 'Traffic Conditions'], drop_first=True)

    # Define the expected column order based on the model's training data
    model_columns = [
        'Distance (km)', 'year',
        'Origin_Ahmedabad', 'Origin_Bangalore', 'Origin_Chennai', 'Origin_Delhi',
        'Origin_Hyderabad', 'Origin_Jaipur', 'Origin_Kolkata', 'Origin_Lucknow',
        'Origin_Mumbai', 'Origin_Pune', 'Destination_Ahmedabad', 'Destination_Bangalore',
        'Destination_Chennai', 'Destination_Delhi', 'Destination_Hyderabad', 'Destination_Jaipur',
        'Destination_Kolkata', 'Destination_Lucknow', 'Destination_Mumbai', 'Destination_Pune',
        'Vehicle Type_Container', 'Vehicle Type_Lorry', 'Vehicle Type_Trailer',
        'Vehicle Type_Truck', 'Weather Conditions_Clear', 'Weather Conditions_Fog',
        'Weather Conditions_Rain', 'Weather Conditions_Storm', 'Traffic Conditions_Heavy',
        'Traffic Conditions_Light', 'Traffic Conditions_Moderate'
    ]

    # Add missing columns with 0 if they are not present after encoding
    missing_cols = set(model_columns) - set(shipment_data.columns)
    for col in missing_cols:
        shipment_data[col] = 0  # Add missing columns with default value of 0

    # Reorder the columns to match the expected order
    shipment_data = shipment_data[model_columns]

    # Convert the final DataFrame to integers (model works better with integer values)
    shipment_data = shipment_data.astype(int)

    # Convert the final DataFrame to a numpy array
    final_features = shipment_data.values

    # Make prediction using the trained AdaBoost model
    prediction = xgb_model.predict(final_features)

    # Return the prediction result (Delayed or On Time)
    if prediction == 1:
        return "Delayed"
    else:
        return "On Time"

# Step 3: Create the API route
@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Get JSON data from the POST request
        data = request.get_json()

        # Ensure that the required fields are provided
        required_fields = ['Shipment ID', 'Origin', 'Destination', 'Shipment Date', 'Vehicle Type', 'Distance (km)', 'Weather Conditions', 'Traffic Conditions']

        if not all(field in data for field in required_fields):
            return jsonify({"error": "Missing required fields in input data"}), 400

        # Get the shipment details
        shipment_details = [
            data['Shipment ID'], data['Origin'], data['Destination'], data['Shipment Date'],
            data['Vehicle Type'], data['Distance (km)'], data['Weather Conditions'], data['Traffic Conditions']
        ]

        # Make the prediction
        prediction = predict_shipment_status(shipment_details)

        # Return the prediction result
        return jsonify({"prediction": prediction}), 200

    except Exception as e:
        return jsonify({"error": str(e)}), 500

# Step 4: Run the Flask app using ngrok
def run():
    app.run(port=5001)

# Start the Flask app in a separate thread
thread = Thread(target=run)
thread.start()

# Create a public URL using ngrok
public_url = ngrok.connect(5001)
print('Public URL:', public_url)
```
